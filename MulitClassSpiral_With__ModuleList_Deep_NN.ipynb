<h2>Preparation</h2>

We'll need the following libraries

```python
# Import the libraries we need for this lab

import matplotlib.pyplot as plt 
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from matplotlib.colors import ListedColormap
from torch.utils.data import Dataset, DataLoader

torch.manual_seed(1)
```

Function used to plot:

```python
# Define the function to plot the diagram

def plot_decision_regions_3class(model, data_set):
    cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#00AAFF'])
    cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#00AAFF'])
    X = data_set.x.numpy()
    y = data_set.y.numpy()
    h = .02 #shifting coeffcient which will used in arrange
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1 
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1 
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    XX = torch.Tensor(np.c_[xx.ravel(), yy.ravel()])# here ravel is return flattned cotagious array
    _, yhat = torch.max(model(XX), 1) #supplying model's output to agmax fn for softmax prediction
    yhat = yhat.numpy().reshape(xx.shape)  #reshape correponds to x's values
    plt.pcolormesh(xx, yy, yhat, cmap=cmap_light)
    plt.plot(X[y[:] == 0, 0], X[y[:] == 0, 1], 'ro', label = 'y=0')#for y=0
    plt.plot(X[y[:] == 1, 0], X[y[:] == 1, 1], 'go', label = 'y=1')#for y=1
    plt.plot(X[y[:] == 2, 0], X[y[:] == 2, 1], 'o', label = 'y=2')#for y=2
    plt.title("decision region")
    plt.legend()
```

Create Dataset <code>Class</code>

```python
# Create Data Class
#The toy spiral data consists of three classes  that are not linearly separable.

class Data(Dataset):
    
    #  modified from: http://cs231n.github.io/neural-networks-case-study/
    # Constructor
    def __init__(self, K=3, N=500):
        D = 2
        #initialized x & y
        X = np.zeros((N * K, D)) # data matrix (each row = single example) ---------there is 1500(as N*K=500*3) pair of zeros
        y = np.zeros(N * K, dtype='uint8') # class labels   here theire is 1500(as N*K=500*3) only single values of zeros 
        for j in range(K):
          ix = range(N * j, N * (j + 1))
          r = np.linspace(0.0, 1, N) # radius
          t = np.linspace(j * 4, (j + 1) * 4, N) + np.random.randn(N) * 0.2 # theta
          X[ix] = np.c_[r * np.sin(t), r*np.cos(t)]
          y[ix] = j
        self.y = torch.from_numpy(y).type(torch.LongTensor)
        self.x = torch.from_numpy(X).type(torch.FloatTensor)
        self.len = y.shape[0]
    
    # Getter
    def __getitem__(self, index):    
        return self.x[index], self.y[index]
    
    # Get Length
    def __len__(self):
        return self.len
    
    # Plot the diagram
    def plot_stuff(self):
        plt.plot(self.x[self.y[:] == 0, 0].numpy(), self.x[self.y[:] == 0, 1].numpy(), 'o', label="y = 0")
        plt.plot(self.x[self.y[:] == 1, 0].numpy(), self.x[self.y[:] == 1, 1].numpy(), 'ro', label="y = 1")
        plt.plot(self.x[self.y[:] == 2, 0].numpy(), self.x[self.y[:] == 2, 1].numpy(), 'go', label="y = 2")
        plt.legend()
```

<h2 id="Model">Neural Network Module and Function for Training</h2>

Neural Network Module using <code>ModuleList()</code>

```python
# Create Net model class

class Net(nn.Module):
    
    # Constructor
    def __init__(self, Layers):
        super(Net, self).__init__()
        self.hidden = nn.ModuleList()#making object from constructor nn.ModuleList
        for input_size, output_size in zip(Layers, Layers[1:]):#Layers, Layers[1:]= current number in layer_list & its next number in layer_list relatively & iteratively
            self.hidden.append(nn.Linear(input_size, output_size)) #then the above numbers with pair(Layers, Layers[1:]) used as input_size(or no.of neurons in previous layer)[if its first hidden layer then this no. used as no. of nuerons in input layer], output_size(or no. of neurons in current layer)
    
    # Prediction
    def forward(self, activation):
        L = len(self.hidden)#length correponds to hidden_layer_list
        for (l, linear_transform) in zip(range(L), self.hidden):
            if l < L - 1: #untill the layer is hidden layer
                activation = F.relu(linear_transform(activation)) #we add linear_tranform for linear regression layer along with activation function(here:RELU)
            else: #for output layer or l=L-1
                activation = linear_transform(activation) #only linear
        return activation
```

A function used to train. 

```python
# Define the function for training the model

def train(data_set, model, criterion, train_loader, optimizer, epochs=100):
    LOSS = []
    ACC = []
    for epoch in range(epochs):
        for x, y in train_loader:
            optimizer.zero_grad()
            yhat = model(x)
            loss = criterion(yhat, y)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            LOSS.append(loss.item())
        ACC.append(accuracy(model, data_set))
    
    fig, ax1 = plt.subplots()
    color = 'tab:red'
    ax1.plot(LOSS, color = color)
    ax1.set_xlabel('Iteration', color = color) #total loss v/s iterations
    ax1.set_ylabel('total loss', color = color)
    ax1.tick_params(axis = 'y', color = color)
    
    ax2 = ax1.twinx()   
    color = 'tab:blue'
    ax2.set_ylabel('accuracy', color = color)  # we already handled the x-label with ax1
    ax2.plot(ACC, color = color)
    ax2.tick_params(axis = 'y', color = color)
    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    
    plt.show()
    return LOSS
```

A function used to calculate accuracy 

```python
# The function to calculate the accuracy

def accuracy(model, data_set):
    _, yhat = torch.max(model(data_set.x), 1)
    return (yhat == data_set.y).numpy().mean()
```

<h2 id="Train">Train and Validate the Model</h2>

Crate a dataset object:

```python
# Create a Dataset object

data_set = Data()
data_set.plot_stuff()
data_set.y = data_set.y.view(-1)
```

Create a  network to classify three classes with 1 hidden layer with 50 neurons 

```python
# Train the model with 1 hidden layer with 50 neurons

Layers = [2, 50, 3] #[inputlayer: 2]-----[hiddenlayer:(2,50)]--------[outputlayer:(50,3)]
model = Net(Layers)  #supply Layer_list to modelfor arbitiary-automation specially for deep neural network 
learning_rate = 0.10
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
train_loader = DataLoader(dataset=data_set, batch_size=20)
criterion = nn.CrossEntropyLoss()
LOSS = train(data_set, model, criterion, train_loader, optimizer, epochs=100)

plot_decision_regions_3class(model, data_set)
```
